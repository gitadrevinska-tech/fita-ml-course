## ML uzdevuma tips

*Šis ir supervised uzdevums, jo mums ir zināms, kas izdzīvoja/neizdzīvoja. Un šī būs klasifikācija, jo modelis pasažieriem piešķirs vienu no divām vērtībām (0 Neizdzīvoja, 1 Izdzīvoja), bet, izmantojot vairāku kolonnu datus, piem., sex, age, sibsp, mācīsies atpazīt, kādiem pasažieriem bija lielāka izdzīvošanas varbūtība, un tāpēc varēspārbaudīt, vai princips "sievietes un bērni vispirms" tiešām tika ievērots.*


```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')

# Pārbaud
df.head()
```

```python
# Target mainīgais
print(df['Survived'].value_counts())
print(f'\nIzdzīvoja: {df["Survived"].mean():.1%}')
```

## Target variable

*Target mainīgais ir tas, ko vēlamies prognozēt. Šajā gadījumā tas ir survived — kolonna, kas norāda, vai pasažieris izdzīvoja Titānika katastrofā.*

## Features saraksts

**Skaitliskās (numerical):**
age -> Vecums gados;
fare -> Biļetes cena;
sibsp -> Brāļu/māsu un laulāto skaits;
parch -> Vecāku/bērnu skaits.

**Kategoriskās (categorical):**
sex -> Dzimums;
pclass -> Biļetes klase (skaitliska);
embarked -> Iekāpšanas osta;
alone -> Vai ceļoja viens;
class -> Biļetes klase (vārdisks).

```python
# Features pārskats
print('Skaitliskās features:')
print(df.select_dtypes(include='number').columns.tolist())
print('\nKategoriskās features:')
print(df.select_dtypes(include='str').columns.tolist())
```
